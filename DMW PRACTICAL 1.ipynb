### 	Then using request library I fetched the raw data of the website(https://www.goodreads.com/author/list/1077326.J_K_Rowling)  :

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data.content, 'html.parser')
print(soup.prettify())
```

### #	Then I further proceeded to expand the tags and get into the particular tag so that I can scrape the required data :

```python
print(soup.title)
```

```python
print(soup.title.name)
```

```python
s=soup.find('table', class_='tableList')
```

```python
print(s)
```

### #	This finds all links `(<a> tags) `within the table with the class 'tableList' that have the class 'bookTitle' :

```python
all_links=s.find_all('a', class_='bookTitle')
```

```python
print(all_links)
```

### 	This iterates through all the book links, extracts the <span> tag with the attribute {'role': 'heading'}, and prints the text content of that span tag :

```python
for link in all_links:
    soup = BeautifulSoup(str(link), 'html.parser')
    span_tag = soup.find('span', {'role': 'heading'})
    print(span_tag.text)
```
