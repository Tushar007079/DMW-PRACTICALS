```python
import io
import pandas as pd
```

```python
from google.colab import files
uploaded =files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['NBA_2018-19_Season - NBA_2018-19_Season.csv']))
```

```python
df
```

```python
#1)Find out the average age of the player who have played for NBA?
df['Age'].mean()
```

```python
#2)Find out that each player have played how many games?
df.groupby('Player')['Games'].max()
```

```python
#3)Total how many teams are there in NBA?
df['Team'].nunique()
```

```python
#4)What is the minimum age of the player that have played for NBA?
df['Age'].min()
```

```python
#5)What is the Maximum age of the player that have played for NBA and display its details?
df[df.Age==df.Age.max()]
```

```python
#6)Find out how many games have been organized in Eastern region.
df[df['Conference']=="Eastern"]['Games'].sum()
```

```python
#7)Find out the how many regions are there, where the games have been organized?
df['Conference'].nunique()
```

```python
#8)Display the list of the player who have played in the team called "Boston Celtics".
df[df.Team=="Boston Celtics"]['Player']
```

```python
#9)Find out the total number of games organized in each division?
df.groupby('Division')['Games'].sum()
```

```python
#10)Which player have scored the maximum number of goals?
df[df['Field Goals']==df['Field Goals'].max()]['Player']
```

```python
#11)Which player have the lowest number of personal fouls?
df[df['Personal Fouls']==df['Personal Fouls'].min()]['Player']
```

```python
#12)Which player have attempted the 3-points fi eld goals the highest time and what are the percentage of it?
df[df['3-Point Field Goal Attempts'] == df['3-Point Field Goal Attempts'].max()][['Player', '3-Point Field Goal Percentage']]
```

```python
#13)What is the average point scored by all the players?
df['Points'].mean()
```

```python
#14)Find out the average age of players division wise?
df.groupby('Division')['Age'].mean()
```

```python
#15)Find out the total number of fouls in each team?
df.groupby("Team")['Personal Fouls'].sum()
```
