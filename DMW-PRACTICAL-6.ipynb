Question 1:

Assume a dataset of 10 transactions which has
the list of the items that have been bought for each transactions. So the dataset will be having two kind of information in the dataset that is – (1) Transaction ID and (2) List of items.

Find the support value of the each combination of the items.


```python
dataset = [['Milk','Onion','Nutmeg','Eggs','Yogurt'],
['Bread','Onion','Nutmeg','Eggs','Yogurt'],
['Milk','Apple','Eggs'],
['Milk','Unicorn','Corn','Yogurt'],
['Corn','Onion','Ice Cream','Eggs'],
['Bread','Onion','Nutmeg','Yogurt'],
['Bread','Apple','Eggs','Yogurt'],
['Milk','Onion','Nutmeg'],
['Corn','Onion','Ice Cream','Eggs'],
['Bread','Milk','Corn']]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
tel=te.fit(dataset).transform(dataset)
```

```python
tel
```

```python
df = pd.DataFrame(tel, columns=te.columns_)
df
```

```python
test=tel[:5]
te.inverse_transform(test)
```

```python
tel.astype('int')
```

```python
te.columns_
```

```python
from mlxtend.frequent_patterns import apriori
```

Find the support value of the each combination of the items.

```python
apriori(df, min_support = 0.1,use_colnames=True)
```

```python
freq = apriori(df,min_support = 0.4, use_colnames=True)
freq['length']=freq['itemsets'].apply(lambda x:len(x))
freq
```

```python
freq[freq['itemsets'].apply(lambda x: 'Yogurt' in x and 'Kidney Beans' in x)]
```

```python
freq[freq['itemsets']=={'Kidney Beans','Eggs'}]
```

```python
freq[(freq['length']==2) & (freq['support']>=0.6)]
```
