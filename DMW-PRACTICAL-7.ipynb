```python
data['Name']=data['Name'].map(str).apply(lambda x:x.encode('utf-8').decode('ascii','ignore'))
data['Boat [Body]']=data['Boat [Body]'].map(str).apply(lambda x:x.encode('utf-8').decode('ascii','ignore'))
data.head(5)
```

```python
import re

def proc_age(value):
    if 'm' in value:
        return float(re.findall(r'-?\d+\.?\d*', value)[0]) / 12
    else:
        return float(value)

data['Age'] = data['Age'].map(str).apply(proc_age)

# Round the Age column to 2 decimal places
data['Age'] = data['Age'].round(2)

# Display the first few rows of the DataFrame
data.head()
```

```python
data_trim = data [['Name','Age','Class/Dept','Boat [Body]']]
data_trim.head()
```

```python
def categories_person(value):
    if 'PASSENGER' in value.upper():
        return 'Passenger'
    else:
        return 'Crew'

data_trim['Crew/Passenger'] = data_trim['Class/Dept'].map(str).apply(categories_person)

# Renaming columns for better clarity
data_trim = data_trim.rename(columns={'Class/Dept': 'Class_or_Dept'})

# Display the first few rows of the DataFrame
data_trim.head()
```

```python
def categories_age(age):
    if age > 18:
        return 'Adult'
    else:
        return 'Child'

data_trim['Adult/Child'] = data['Age'].apply(categories_age)

# Placeholder for formatting
# Apply formatting actions here if needed

data_trim.head()
```

```python
def categories_gender(name):
    firstname = name[name.index(',') + 2:]
    salutation = firstname.split(" ")[0]
    if salutation.upper() in ['MR', 'MASTER']:
        return 'Male'
    else:
        return 'Female'

data_trim['Gender'] = data_trim['Name'].map(str).apply(categories_gender)

# Placeholder for formatting
# Apply formatting actions here if needed

data_trim.head()
```

```python
def categories_survival(value):
    if value.strip() == 'nan' or '[' in value:
        return 0  # Not survived
    else:
        return 1  # Survived

data_trim['Survival'] = data_trim['Boat [Body]'].map(str).apply(categories_survival)

# Placeholder for formatting
# Apply formatting actions here if needed

data_trim.head()
```

```python
data_trim.groupby(['Crew/Passenger'])['Survival'].sum()*100/data_trim.groupby(['Crew/Passenger'])['Survival'].count()
print("\n\n\n")
data_trim.groupby(['Adult/Child'])['Survival'].sum()*100/data_trim.groupby(['Adult/Child'])['Survival'].count()
```

```python
def categories_age(age):
    if age <= 13:
        return '0-13'
    elif 14 <= age <= 18:
        return '14-18'
    elif 19 <= age <= 30:
        return '19-30'
    elif 31 <= age <= 50:
        return '31-50'
    else:
        return '50 and above'

data_trim['Age_classification'] = data['Age'].apply(categories_age)

# Placeholder for formatting
# Apply formatting actions here if needed

data_trim.head()
```
